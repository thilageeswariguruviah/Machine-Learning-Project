```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['title'] = label_encoder.fit_transform(data['title'])
data['uploaded_by'] = label_encoder.fit_transform(data['uploaded_by'])
data['last_updated'] = label_encoder.fit_transform(data['last_updated'])
data['usability'] = label_encoder.fit_transform(data['usability'])
data['files'] = label_encoder.fit_transform(data['files'])
data['size'] = label_encoder.fit_transform(data['size'])
data['upvotes'] = label_encoder.fit_transform(data['upvotes'])
data['badge'] = label_encoder.fit_transform(data['badge'])
```

```python
X = data.iloc[:,1:-1]#independent variable
Y = data.iloc[:, -1]#dependent variable
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.25, random_state=7)
```

```python
X_test
```

```python
Y_train
```

Decision tree classifier

```python
# Import decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1234)
dtc.fit(X_train, Y_train)
```

```python
#Predicting the test set result
y_predict = dtc.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_predict)
score = dtc.score(X_test,Y_test)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score
accuracy_dtc = accuracy_score(Y_test,y_predict)
print('Accuracy',accuracy_dtc)
recall = recall_score(Y_test,y_predict,average = 'macro')
print('Recall',recall)
f1_score = f1_score(Y_test,y_predict,average ='macro')
print('data1 score',f1_score)
```

KNN Classification

```python
# Import KNN_classification
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=5,
                       p=2 )
classifier.fit(X_train, Y_train)
```

```python
#Predicting-1 the test set result
y1_pred= classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y1_pred)
score = dtc.score(X_test,Y_test)
```
